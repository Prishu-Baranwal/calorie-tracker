```python
dataset_path = r"C:\Users\Prishu Baranwal\dataset\food-101\food-101\images"

# Remove .DS_Store from categories (it's a Mac system file, not a food)
categories = [c for c in os.listdir(dataset_path) if c != '.DS_Store']
print(f"Total food categories: {len(categories)}")
print(f"Sample categories: {categories[:10]}")
```

```python
sample_food = categories[0]
sample_img_path = os.path.join(dataset_path, sample_food)
sample_img = os.listdir(sample_img_path)[0]

img = Image.open(os.path.join(sample_img_path, sample_img))
plt.imshow(img)
plt.title(f"Sample Food: {sample_food}")
plt.axis('off')
plt.show()
```

```python
import urllib.request

# Download nutrition CSV directly (no Kaggle needed)
url = "https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/nutrients.csv"
save_path = r"C:\Users\Prishu Baranwal\dataset\nutrition.csv"

print("Downloading nutrition dataset...")
urllib.request.urlretrieve(url, save_path)
print("✅ Downloaded successfully!")

# Load and check it
nutrition_df = pd.read_csv(save_path)
print(nutrition_df.head())
print(f"\nTotal food items in nutrition data: {len(nutrition_df)}")
print(f"Columns: {list(nutrition_df.columns)}")
```
